```python
# Libraries used:

import shutil
import random
import os
```

```python
russian_ds = os.listdir("cat_dog_breed/images/images/")

required_breeds = ["Siamese", "Domestic Long Hair", 
                   "Domestic Short Hair", 
                   "Bengal", 
                   "Bombay", 
                   "Russian Blue Cat"] # sets a placeholder for the breeds required
catbreeds_ds = os.listdir("cat_ds/images/")
```

```python
# classifies between all the breeds in the different folders 

for breed in catbreeds_ds:
    if breed in required_breeds:
        breed_images = os.listdir(f'cat_ds/images/{breed}')
        for img in breed_images:
            print(img)
            shutil.copyfile(f"cat_ds/images/{breed}/{img}", f"cat_breed/{breed}/{img}") 
```

```python
for file in russian_ds:
    if "Russian_Blue" in file:
        shutil.copy(f"cat_dog_breed/images/images/{file}", f"cat_breed/Russian Blue Cat/{file}")
```

```python
# renaming all files to relate them to their breeds

for breed in required_breeds:
    tally = 1
    breed_images = os.listdir(f'data/cat_breed/{breed}/')
    for file in breed_images:
        os.rename(f"data/cat_breed/{breed}/{file}", f"data/cat_breed/{breed}/{breed}_{tally}.jpg")
        tally += 1
    
```

```python
# inflating cat set with data augmentation

from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img


datagen = ImageDataGenerator(
        rotation_range = 10,
        shear_range = 0.1,
        zoom_range = 0.125,
        horizontal_flip = True,
        brightness_range = (0.75, 2.0))


for f in os.listdir(f'data/cat_breed/Russian Blue Cat'):
    img = load_img(f"data/cat_breed/Russian Blue Cat/{f}")  
    x = img_to_array(img)
    x = x.reshape((1, ) + x.shape)  
    i = 0

    # generate 1 augmented image for each original image
    for batch in datagen.flow(x, batch_size = 1, 
                      save_to_dir =f'data/cat_breed/{breed}',  
                      save_prefix =f"{breed}_d_", save_format ='jpg'): 
        i += 1
        if i > 1: 
            break
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import os
import time
import urllib.request
from selenium.webdriver.common.keys import Keys

# using drivers to count the number of images

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.google.com/')
search = driver.find_element("name", 'q')    
search.send_keys(f'Russian Blue Cat breed images',Keys.ENTER)

elem = driver.find_element("link text", 'Images')
elem.get_attribute('href')
elem.click()

value = 0
for i in range(70):

    print(f"{breed} load progress -> {i}")
    driver.execute_script('scrollBy("+ str(value) +",+500);')
    value += 500
    time.sleep(0.420)

elements = driver.find_elements('xpath', '//img[contains(@class,"rg_i")]')
count = 0
for i in elements:
    src = i.get_attribute('src')
    try:
        if src != None:
            src = str(src)
            count += 1
            print(f'img count -> {count}')
            urllib.request.urlretrieve(src, os.path.join(f'data/cat_breed/Russian Blue Cat/',f'{breed}_g'+str(count)+'.jpg'))
        else:
            raise TypeError
    except TypeError:	
        pass
```
